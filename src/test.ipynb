```python
import torch
from torch import nn
device = "cpu"
if torch.cuda.is_available():
    device = "cuda"
```

```python
torch.set_default_device(device)
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.w = nn.Parameter(torch.tensor([[1,2],[2,2]],dtype=torch.float16))
    def forward(self,x):
        return self.w*x
m = Model()
result = m(10).reshape(-1,1)
loss = nn.L1Loss()(result,torch.randint(1,2,(4,1),dtype=torch.float16))
```
